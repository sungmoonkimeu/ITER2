## Operations

```python
# Sum
M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1+J2)

M1 = np.ones((2,2,3))
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2,3))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1+J2)

M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2,3))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1+J2)

M1 = np.ones((2,2,6))
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2,3, 2))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1+J2)

try:
    print(J1+1)
except:
    print('Funciona')
```

```python
# Substraction
M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1-J2)

M1 = np.ones((2,2,3))
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2,3))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1-J2)

M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2,3))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1-J2)

M1 = np.ones((2,2,6))
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2,2,3, 2))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1-J2)

try:
    print(J1-1)
except:
    print('Funciona')
```

```python
# Left multiplication by constants
M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
print(J1*2)

M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
a = np.linspace(1, 5, 5)
print(J1*a)

M1 = np.ones((2, 2, 5))
J1 = Jones_matrix()
J1.from_matrix(M1)
a = np.linspace(1, 5, 5)
print(J1*a)
```

```python
# Left multiplication by Jones vectors
M = [2, 3j]
E = Jones_vector()
E.from_matrix(M)
M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
print(J1 * E)


a = np.linspace(0, 90*degrees, 5)
E = Jones_vector()
E.linear_light(azimuth=a)
print(E)
M1 = [[1, 2],[3, 4]]
J1 = Jones_matrix()
J1.from_matrix(M1)
print(J1 * E)

M = [2, 3j]
E = Jones_vector()
E.from_matrix(M)
M1 = np.ones((2, 2, 5))
J1 = Jones_matrix()
J1.from_matrix(M1)
print(J1 * E)

a = np.linspace(0, 90*degrees, 5)
E = Jones_vector()
E.linear_light(azimuth=a)
print(E)
M1 = np.ones((2, 2, 5))
J1 = Jones_matrix()
J1.from_matrix(M1)
print(J1 * E)
```

```python
# Left multiplication by Jones matrices
M1 = [[1, 2], [3, 4]]
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.eye(2)
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1 * J2)


M1 = [[1, 2], [3, 4]]
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2, 2, 4))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1 * J2)

M1 = np.ones((2, 2, 4))
J1 = Jones_matrix()
J1.from_matrix(M1)
M2 = np.ones((2, 2, 4))
J2 = Jones_matrix()
J2.from_matrix(M2)
print(J1 * J2)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(a@b)

J1, J2 = create_Jones_matrices(N=2)
J1.from_matrix(a)
J2.from_matrix(b)
print(J1*J2)
```

```python
a2 = np.zeros((2,2,2))
a2[:,:,0] = a
b2 = np.zeros((2,2,2))
b2[:,:,0] = b
print(a2@b2)

J1, J2 = create_Jones_matrices(('J1', 'J2'))
J1.from_matrix(a2)
J2.from_matrix(b2)
print(J1, J2)
print(J1*J2)
```

```python
# Right multiplication by constants
M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
print(2*J1)

M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
a = np.linspace(1, 5, 5)
print(a*J1)

M1 = np.ones((2, 2, 5))
J1 = Jones_matrix()
J1.from_matrix(M1)
a = np.linspace(1, 5, 5)
print(a*J1)
```

```python
# Division
M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
print(J1/2)

M1 = np.eye(2)
J1 = Jones_matrix()
J1.from_matrix(M1)
a = np.linspace(1, 5, 5)
print(J1/a)

M1 = np.ones((2, 2, 5))
J1 = Jones_matrix()
J1.from_matrix(M1)
a = np.linspace(1, 5, 5)
print(J1/a)
```

```python
# Rotation
M1 = [[1, 0],[0, 0]]
J1 = Jones_matrix()
J1.from_matrix(M1)
J1.rotate(45*degrees)
print(J1)


M1 = [[1, 0],[0, 0]]
J1 = Jones_matrix('J1')
J1.from_matrix(M1)
angles = np.linspace(0, 180*degrees, 11)
J2 = J1.rotate(angles)
J2.name = 'J2'
print(J1)
print(J2)

M1 = [[1, 0],[0, 0]]
J1 = Jones_matrix('J1')
J1.from_matrix(M1)
angles = np.linspace(0, 90*degrees, 11)
J2 = J1.rotate(angles, keep=True)
J2.name = 'J2'
print(J1)
print(J2)

M1 = np.ones((2, 2, 11))
J1 = Jones_matrix('J1')
J1.from_matrix(M1)
angles = np.linspace(0, 90*degrees, 11)
J1.rotate(angles)
print(J1)


M1 = np.ones((2, 2, 11))
J1 = Jones_matrix('J1')
J1.from_matrix(M1)
angles = 45*degrees
J1.rotate(angles)
print(J1)
```

```python
# Reciprocal element
M1 = np.random.rand(2, 2, 5) + 1j * np.random.rand(2, 2, 5)
J1 = Jones_matrix('J1')
J1.from_matrix(M1)
print(J1)
print(J1.reciprocal())
```

## Manipulations

```python
# Sum
M = np.random.rand(2,2,5)
M2 = np.sum(M, axis=2)
print(M2)
J = Jones_matrix('J1')
J.from_matrix(M)
J1 = J.sum(axis=0)
print(J1, '\n\n')

M = np.random.rand(2,2, 2,2)
M2 = np.sum(M, axis=2)
print(M2)
J = Jones_matrix('J1')
J.from_matrix(M)
J1 = J.sum(axis=0)
print(J1, '\n\n')

M = np.random.rand(2,2, 2,2, 2)
M2 = np.sum(M, axis=(2,3))
print(M2)
J = Jones_matrix('J1')
J.from_matrix(M)
J1 = J.sum(axis=[0,1])
print(J1, '\n\n')
```

```python
# prod
M = np.random.rand(2,2,3)
J = Jones_matrix('J')
J.from_matrix(M)
J1 = J.prod()
print(J1, '\n\n')

M = np.random.rand(2,2,2, 3)
J = Jones_matrix('J')
J.from_matrix(M)
J1 = J.prod(axis=0)
print(J1, '\n\n')

M = np.random.rand(2,2,2, 3)
J = Jones_matrix('J')
J.from_matrix(M)
J1 = J.prod(axis=[0, 1])
print(J1, '\n\n')

M = np.random.rand(2,2,2, 3, 4)
J = Jones_matrix('J')
J.from_matrix(M)
J1 = J.prod(axis=1)
print(J1, '\n\n')

M = np.random.rand(2,2,2, 3, 4)
J = Jones_matrix('J')
J.from_matrix(M)
J1 = J.prod(axis=[0, 1])
print(J1, '\n\n')
```

```python
# Rotation
M = np.array([[1, 1], [1, 0]])
angle = [0*degrees, 45*degrees, 90*degrees, 120*degrees, 135*degrees]
J = Jones_matrix('J1')
J.from_matrix(M)
J.rotate(angle, change_name=False)
print(J)


angle = 20*degrees
J.from_matrix(M)
J2 = J.rotate(angle, change_name=False, keep=True)
J2.name = 'J2'
print(J2, '\n', J)
```

```python
# getitem
M = np.random.rand(2, 2, 3, 5)
J = Jones_matrix('Test')
J.from_matrix(M)
print(J)

J2 = J[0]
print(J2)

J2 = J[0:3]
print(J2)

J3 = J[1, 1]
print(J3)

J3 = J[0:3, 0:2]
print(J3)

J4 = J[np.linspace(2, 6, 3, dtype=int)]
print(J4)

aux = M[0,0,:,:] > 0.5
J4 = J[aux]
print(J4)
```

```python
# setitem
M = np.random.rand(2, 2, 3, 5)
J = Jones_matrix('Test')
J2 = Jones_matrix('New')
J.from_matrix(M)
print(J)

J2.vacuum()
J[0] = J2
print(J)

angle = np.array([0, 45, 90]) * degrees
J2.diattenuator_linear(p1=1, p2=0, azimuth=angle)
J.from_matrix(M)
J[0:3] = J2
print(J)

J2.vacuum()
J.from_matrix(M)
J[1,1] = J2
print(J)

J2.vacuum(length=6)
J2.shape = [3, 2]
J.from_matrix(M)
J[0:3, 0:2] = J2
print(J)

J2.vacuum(length=3)
J.from_matrix(M)
J[np.linspace(2, 6, 3, dtype=int)] = J2
print(J)

J2.vacuum()
aux = M[0,0,:,:] > 0.5
J.from_matrix(M)
J[aux] = J2
print(J)
```
