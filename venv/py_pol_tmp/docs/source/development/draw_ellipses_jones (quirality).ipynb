# Drawing ellipses for Jones vectors

https://en.wikipedia.org/wiki/Stokes_parameters

```python
%matplotlib notebook 
from py_pol import degrees, np
from py_pol.jones_vector import Jones_vector
import matplotlib.pyplot as plt
from py_pol.drawings import draw_ellipse_jones, draw_ellipses_jones
```

## Only 1 vector - for Jones_vector class

```python
j0s=Jones_vector('j0')
j0s.circular_light(amplitude=1, kind='r')
j0s.elliptical_light(a=10, b=20, phase=45*degrees, angle=45*degrees)
```

```python
j0s.draw_ellipse(draw_arrow=True)
```

```python
Jones_vectors_0=[]

j0=Jones_vector('j0')
j0.linear_light(amplitude=1, angle=0*degrees)


angles = np.linspace(0, 180*degrees, 6, endpoint=False)

for i, angle in enumerate(angles):
    ji=j0.rotate(angle=angle, keep=False)
    Jones_vectors_0.append(ji)
    print(ji)
    
print (Jones_vectors_0)
```

```python
Jones_vectors_1=[]

j0=Jones_vector('j0')
j0.elliptical_light(a=20, b=20, phase=45*degrees, angle=0*degrees)


angles = np.linspace(0, 180*degrees, 6, endpoint=False)

for i, angle in enumerate(angles):
    ji=j0.rotate(angle=angle, keep=False)
    Jones_vectors_1.append(ji)
    print(ji)
    
print (Jones_vectors_1)
```

```python
ax, fig=draw_ellipses_jones(Jones_vectors_0, filename='', draw_arrow=True)
```

```python
ax, fig=draw_ellipses_jones(Jones_vectors_1, filename='', draw_arrow=True)
```
