```python
# Polarizer right
N = 20
errors = np.zeros(N)
for ind in range(N):
    # Create random parameters
    (p1a, p2a, alphaP, delayP, delta, alphaR, delayR) = np.random.rand(7)
    p1 = max(p1a, p2a)
    p2 = min(p1a, p2a)
    alphaP = alphaP*90*degrees
    delayP = delayP*360*degrees
    delta = delta*180*degrees
    alphaR = alphaR*90*degrees
    delayR = delayR*360*degrees     
#     solution = (p1, p2, alphaP, delayP, delta, alphaR, delayR)
    # Create objects 
    M1 = Mueller('Mp')
    M1.diattenuator_charac_angles_from_Jones(p1, p2, alphaP, delayP)
    M2 = Mueller('Mr')
    M2.retarder_charac_angles_from_Jones(delta, alphaR, delayR)
    M = M2*M1
    # Decompose
    Mr, Md, params = M.analysis.decompose_pure(give_all = True, tol=1e-10)
    aux = [comparison(Md.M, M1.M, eps), comparison(Mr.M, M2.M, eps)]
#     print(aux)   
#     print([delta, alphaR, delayR])
#     print(Mr.analysis.retarder())
#     print("")
#     print([Mr, M2])
#     print([Md, M1])
    errors[ind] = all(aux)
print(errors)
Mr, Md = M.analysis.decompose_pure(tol=1e-10, verbose=True)
```
